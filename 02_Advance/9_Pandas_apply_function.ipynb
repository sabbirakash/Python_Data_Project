```python
df1.dropna(subset = "salary_year_avg", inplace = True)
```

```python
df1["salary_year_inflated"] = df1["salary_year_avg"].apply(lambda x : x * 1.03)
df1[["salary_year_avg", "salary_year_inflated"]]
```

```python
def senior_junior_salary(row):
    if "Senior" in row["job_title_short"]:
        return row["salary_year_avg"] * 1.05
    else :
        return row["salary_year_avg"] * 1.03
```

```python
df1["salary_year_avg_inflated"] = df1.apply(senior_junior_salary, axis = 1)
df1[["job_title_short","salary_year_avg", "salary_year_avg_inflated"]]
```

Same thing using lambda function

```python
df1["salary_year_avg_inflated2"] = df1.apply(lambda x : x["salary_year_avg"] * 1.05 if "Senior" in x["job_title_short"] else x["salary_year_avg"] * 1.03, axis = 1)
```
