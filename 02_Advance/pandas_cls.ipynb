### Advance Analysis using GroupBy

```python
df.groupby('job_title_short')['salary_year_avg'].min()
```

```python
df.info()
```

```python
df.groupby(["job_title_short", "job_country"])["salary_year_avg"].min()
```

```python
df.groupby(["job_title_short", "job_country"])["salary_year_avg"].agg([min, max])
```

```python
df["job_country"].isin(["India"]).any()
```

```python
df["job_country"].isin(["North Korea"]).any()
```

```python
us_jobs = df[df["job_country"] == "United States"]
```

```python
us_jobs = us_jobs[us_jobs['salary_year_avg'].notna()]
```

```python
us_jobs
```

```python
us_jobs.groupby("job_title_short")["salary_year_avg"].agg(['median', 'min', 'max', 'count']).sort_values(by = 'median', ascending = False)
```

```python
df.groupby("job_title_short")["salary_year_avg"].median().sort_values(ascending = False)
```
