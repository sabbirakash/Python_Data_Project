```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

# Data Cleanup
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
df_usa = df[df["job_country"] ==  "United States"].copy()
df_usa
```

```python
df_usa["job_posted_month"] = df_usa["job_posted_date"].dt.strftime("%B")

df_usa
```

```python
df_pivot = df_usa.pivot_table(index = "job_posted_month", columns = "job_title_short", aggfunc = "size")
df_pivot
```

```python
df_pivot.reset_index(inplace = True)
df_pivot
```

```python
df_pivot["month_no"] = pd.to_datetime(df_pivot["job_posted_month"], format = "%B").dt.month
df_pivot
```

```python
df_pivot.sort_values("month_no", inplace = True)
df_pivot
```

```python
df_pivot.set_index("job_posted_month", inplace = True)
df_pivot
```

```python
df_pivot.drop(columns = ["index", "month_no"], inplace = True)
df_pivot
```

```python
top_3_jobs = df_usa["job_title_short"].value_counts().head(3).index.to_list()
top_3_jobs
```

```python
df_pivot[top_3_jobs].plot(kind = "line")
plt.title("Monthly Job posting for Top Data Jobs in US")
plt.xlabel("2023")
plt.ylabel("Job Count")
plt.legend()
plt.show()
```

```python
df_pivot_soft = pd.read_csv("https://lukeb.co/software_csv", index_col = "job_posted_month")
df_pivot_soft
```

```python
df_pivot_merged = df_pivot.merge(df_pivot_soft, on = "job_posted_month")
df_pivot_merged
```

```python
top_5_jobs = (df_pivot_merged.sum()     # sum all the jobs of each columns
              .sort_values(ascending=False)         # sorted all the jobs with the highest counts
              .head()               # for the top 5 records
              .index            # from series to a index array
              .to_list())           # from index array to a list
top_5_jobs
```

```python
df_pivot_merged[top_5_jobs].plot(kind = "line")
plt.title("Monthly Job postings for Top Tech Jobs in US")
plt.xlabel("2023")
plt.ylabel("Job Counts")
plt.ylim(0, 20000)
plt.legend()
plt.show()
```
