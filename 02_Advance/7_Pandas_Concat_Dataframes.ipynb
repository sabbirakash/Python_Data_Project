```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

# Data Cleanup
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
df["job_posted_month"] = df["job_posted_date"].dt.strftime("%b")

df.head()
```

```python
months = df["job_posted_month"].unique()
months
```

```python
dict_months = {months : df[df["job_posted_month"] == months] for months in months}
dict_months
```

```python
dict_months["Jan"]
```

```python
dict_months["Dec"]
```

```python
q1_data = pd.concat([dict_months["Jan"], dict_months["Feb"], dict_months["Mar"]], ignore_index = True)
q1_data
```

```python
q1_data["job_posted_month"].value_counts().plot(kind = "bar")
plt.title("Job posting data for first quarter in 2023")
plt.xlabel("2023")
plt.ylabel("Job Count")
plt.show()
```
