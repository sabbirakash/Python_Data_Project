```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

# Data Cleanup
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
df.index
```

```python
df.index.name = "job_index"
df.head()
```

```python
df_usa_job = df[df["job_country"] == "United States"]
df_usa_job.head()
```

```python
df_usa_job.reset_index(inplace = True)

df_usa_job.head()
```

set_index()

```python
df_usa_job.set_index("job_index", inplace = True)
df_usa_job.head()
```

sort_index()

```python
table_pivot = df_usa_job.pivot_table(values = "salary_year_avg", index = "job_title_short", aggfunc = ["median", "min", "max"])
table_pivot
```

```python
table_pivot.sort_values(by = ("median", "salary_year_avg"), ascending = False)
```

```python
table_pivot.sort_index(inplace = True)
table_pivot
```
