```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset("lukebarousse/data_jobs")
df_original = dataset['train'].to_pandas()

# Data Cleanup
df_original["job_posted_date"] = pd.to_datetime(df_original["job_posted_date"])
```

```python
df_original.sample(10, random_state = 31)
```

```python
# Both Dataframe Referencing the same id that's why if one Dataframe is changed the second one is get changed automatically.
df_altered = df_original
```

```python
print("df_original id is :", id(df_original))
print("df_altered id is :", id(df_altered))
print(id(df_original) == id(df_altered))
```
