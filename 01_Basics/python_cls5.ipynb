***Scopes and Closure***

```python
def f1():
  x = 88
  def f2():
    print(x)
  f2()
f1()
```

```python
### Closure
def m1():
  i = 55
  def m2():
    print(i)
  return m2
```

```python
my_result = m1()
my_result()
```

```python
def coderabs(num):
  def actual(x):
    return x ** num
  return actual
```

```python
square_eq = coderabs(2)
square_eq(5)
```

```python
cube_eq = coderabs(3)
cube_eq(5)
```
